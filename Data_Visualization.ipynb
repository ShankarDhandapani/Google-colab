# <h1 align=center><font size = 5>Introduction to Matplotlib </font></h1>

```
#@title Importing Package
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
mpl.style.use('ggplot') # optional: for ggplot-like style
```

```
#@title Reading CSV
f = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/labs/top_selling_albums.csv"
df = pd.read_csv(f)
df
```

```
x = df[['Released','Claimed Sales (millions)']]
x.set_index('Released', inplace=True)
x
```

## <h1 align=center><font size = 5>Line plots </font></h1>

```
x.plot(kind='line')

plt.title('Music Albums')
plt.ylabel('Sales(millions)')
plt.xlabel('Years')
plt.show() # need this line to show the updates made to the figure
```

```
df = pd.read_csv('rainfall in india 1901-2015.csv')
df.head()
```

```
rain = df[['SUBDIVISION', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].groupby("SUBDIVISION").mean()
rain.head()
```

```
rain.plot(kind='area', 
             stacked=True,
             figsize=(20, 10), # pass a tuple (x, y) size
             )

plt.title('India\'s Rainfall between 1901-2015')

plt.show()
```

```
df.hist(figsize=(12,12));
```

```
rain['JAN'].hist(figsize=(8,5));
```

## <h1 align=center><font size = 5>Bar plots </font></h1>

```
rain.plot(kind='bar', figsize=(15, 10))

plt.xlabel('State') # add to x-label to the plot
plt.ylabel('Rainfall') # add y-label to the plot
plt.show()
```

```
# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
df = pd.read_csv('test.csv')
df.head()
```

```
stopwords = set(STOPWORDS)
```

```
# instantiate a word cloud object
wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
wc.generate(str(df['description_x']))
```

```
# display the word cloud
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```
df = pd.read_csv('sample_data/california_housing_train.csv')
df.head()
```

```
# get the first 100 crimes in the df_incidents dataframe
limit = 100
df = df.iloc[0:limit, :]
df.shape
```

```
latitude = 36.778259
longitude = -119.417931 
```

```
# create map and display it
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=5)

# display the map of San Francisco
sanfran_map
```

```
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df.latitude, df.longitude):
    incidents.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow'
        )
    )

# add incidents to map
sanfran_map.add_child(incidents)
```

```
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df.latitude, df.longitude):
    incidents.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow'
        )
    )

# add pop-up text to each marker on the map
latitudes = list(df.latitude)
longitudes = list(df.longitude)

for lat, lng in zip(latitudes, longitudes):
    folium.Marker([lat, lng]).add_to(sanfran_map)    
    
# add incidents to map
sanfran_map.add_child(incidents)
```
